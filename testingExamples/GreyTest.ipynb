# Import Libraries

```python
import sys
sys.path.insert(0,'../')
import matplotlib.pyplot as plt
import ImageOps
imageOps = ImageOps.Ops('grey')
```

Loading a 2 sample images

```python
images = imageOps.convertGrayscale(imageOps.readImage(('image_01grey.jpg','image_02grey.jpg')))
print("images[0] original shape : ", images[0].shape)
print("images[1] original shape : ", images[1].shape)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(images[0], cmap='gray')
ax[0].imshow(images[1], cmap='gray')
```

# Let's resize a images

```python
resizedImage = imageOps.resizeImage(images,140,100)
images = resizedImage
print("images[0] new size shape : ", resizedImage[0].shape)
print("images[1] new size shape : ", resizedImage[1].shape)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(resizedImage[0], cmap='gray')
ax[0].imshow(resizedImage[1], cmap='gray')
```

# Vertical Flip

```python
vFlipImages = imageOps.verticalFlip(images)
print("images[0] flipped image shape : ", vFlipImages[0].shape)
print("images[1] flipped image shape : ", vFlipImages[1].shape)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(vFlipImages[0], cmap='gray')
ax[0].imshow(vFlipImages[1], cmap='gray')
```

# Horizontal Flip

```python
hFlipImages = imageOps.horizontalFlip(images)
print("images[0] flipped image shape : ", hFlipImages[0].shape)
print("images[1] flipped image shape : ", hFlipImages[1].shape)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(hFlipImages[0], cmap='gray')
ax[0].imshow(hFlipImages[1], cmap='gray')
```

# Transpose Flip

```python
hFlipImages = imageOps.transposeFlip(images)
print("images[0] flipped image shape : ", hFlipImages[0].shape)
print("images[1] flipped image shape : ", hFlipImages[1].shape)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(hFlipImages[0], cmap='gray')
ax[0].imshow(hFlipImages[1], cmap='gray')
```

```python
shearImages = imageOps.shear(images,3)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), dpi=80, sharex=True, sharey=True)
ax[1].imshow(shearImages[0], cmap='gray')
ax[0].imshow(shearImages[1], cmap='gray')
```
